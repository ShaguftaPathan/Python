```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"pathname")
df.head()
```

```python
df1 = df.groupby("Traffic Source")["Conversion Rate"].mean().round(2).sort_values(ascending=False)
df1
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
plt.figure(figsize =(10,6))
ax = sns.barplot(x=df1, y=df1.index, palette = "viridis")

plt.title("Avg Conversation Rate by traffic Source")
plt.xlabel("Conversation Rate")
plt.ylabel("Traffic Source")

for bars in ax.containers:
    ax.bar_label(bars);
```

```python
df.groupby("Landing Page")["Bounce Rate"].mean().sort_values(ascending=False)
```

```python


# Group by 'Landing Page' and calculate the mean of 'Bounce Rate'
df2 = df.groupby("Landing Page")["Bounce Rate"].mean().round(2).sort_values(ascending=False)

# Plotting a horizontal bar chart
plt.figure(figsize=(4, 4))
ax = sns.barplot(x=df2.index, y=df2, palette="viridis")

# Adding title and labels
plt.title("Average Bounce Rate by Landing Page")
plt.xlabel("Bounce Rate (%)")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Landing Page")

# Adding the value labels on top of the bars
for bars in ax.containers:
    ax.bar_label(bars)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'Landing Page' and calculate the mean of 'Bounce Rate'
df2 = df.groupby("Landing Page")["Bounce Rate"].mean().round(2).sort_values(ascending=False)

# Plotting a horizontal bar chart
plt.figure(figsize=(6, 3))  # Adjusted size for better visibility
ax = sns.barplot(x=df2, y=df2.index, palette="viridis")

# Adding title and labels
plt.title("Average Bounce Rate by Landing Page")
plt.xlabel("Bounce Rate (%)")
plt.ylabel("Landing Page")

# Rotating x-axis labels to prevent overlap
plt.xticks(rotation=45, ha='right')

# Adding the value labels on top of the bars
for bars in ax.containers:
    ax.bar_label(bars)

# Adjust the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
